```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/NLP Chat Analysis/Dataset/Dataset 3.csv')
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.head()
```

**Store All The Rows From DF With Media into DF2**

```python
df2 = df[df['File'] == "<Media omitted>"]
```

```python
df2.head()
```

**Filter Rows In DF2 Such That Only Rows With Both Message and Media Exist**

```python
df2 = df2.dropna(subset=['Message'])
```

```python
df2.head()
```

**Drop Message Column**

```python
df2 = df2.drop(['Message'],axis=1)
```

```python
df2.head()
```

**Rename File Column as Message Column**

```python
df2.rename(columns = {'File':'Message'}, inplace = True)
```

```python
df2.head()
```

**Change Order Of Columns**

```python
df2 = df2.iloc[:, [0,1,3,2]]
```

```python
df2.head()
```

**Append DF2 Into DF**

```python
df = df.append(df2)
```

```python
df.tail()
```

**Remove Rows Where Both Message and File Exist**

```python
filter_rows = df[pd.notnull(df["Message"]) & pd.notnull(df["File"])].index
```

```python
df = df.drop(filter_rows, axis=0)
```

```python
df.head()
```

**Take (Remaining) Rows With Attachments**

```python
df2 = df[df['File'] == "<Media omitted>"]
```

**Drop (Blank) Message Column**

```python
df2 = df2.drop(['Message'],axis=1)
```

**Rename File as Message and Reorder Columns**

```python
df2.rename(columns = {'File':'Message'}, inplace = True)
```

```python
df2 = df2.iloc[:, [0,1,3,2]]
```

```python
df2.head()
```

**Drop File Column From DF**

```python
df = df.drop(['File'],axis=1)
```

```python
df.head()
```

**Finally Append DF2 into DF**

```python
df = df.append(df2)
```
