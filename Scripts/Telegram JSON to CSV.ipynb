**Read JSON File**

```python
import pandas as pd
df = pd.read_json (r"Dataset 1.json")
```

```python
df.head()
```

**Format Data Into Table**

```python
df = pd.json_normalize(df.messages)
```

```python
df.head()
```

**Drop Unnecessary Columns**

```python
df = df.drop(['id', 'actor', 'actor_id','action',	'title', 
              'from_id',	'width', 'height', 'members',	
              'reply_to_message_id','edited',	'thumbnail', 
              'media_type', 'sticker_emoji',	'mime_type',	
              'message_id','forwarded_from',	'poll.question', 
              'poll.closed', 'poll.total_voters',	'poll.answers',	
              'via_bot','duration_seconds','photo','saved_from','duration',
              'performer', 'game_message_id','score', 'location_information.latitude',	
              'location_information.longitude'], axis=1, errors='ignore')
```

```python
df.head()
```

**Removing Certain Types Of Messages**

```python
patternDel = "^\["
filter = df['text'].str.contains(patternDel,na=True)
```

```python
df = df[~filter]
```

```python
df.head()
```

**Change Media File Syntax To Match With Whatsapp Chat Dataset**

```python
df['file'] = df['file'].replace(['(File not included. Change data exporting settings to download.)'],'<Media omitted>')
```

```python
df.head()
```

**Delete Service Messages**

```python
df.drop(df.loc[df['type']=='service'].index, inplace=True)
```

```python
df.head()
```

**Drop *type* Column**

```python
df = df.drop(['type'],axis=1)
```

```python
df.head()
```

**Rename Columns**

```python
df.columns = ['timestamp', 'message', 'messenger','file']
```

```python
df.head()
```

**Spliting Timestamp into Time and Date**

```python
df[['Date', 'Time']] = df['timestamp'].str.split('T', expand=True)
```

```python
df=df.drop(['timestamp'],axis=1)
```

```python
df = df.iloc[:, [3,4,0,1,2]]
```

```python
df.head()
```

**Export it as a CSV File**

```python
df.to_csv('Dataset 1.csv')
```
